```python
import os
from datetime import datetime

print("Environment Ready.")
print("Welcome to the personal journal manager!")
print("please select an option:")
```

```python
class JournalManager:
    def __init__(self, filename="journal.txt"):
        self.filename = filename
        print(f"Journal initialized. Saving to: {self.filename}")
```

```python
def add_entry(self):
    entry_text = input("\nWhat's on your mind today?\n> ")
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    formatted_entry = f"[{timestamp}]\n{entry_text}\n{'-'*30}\n"
    
    try:
        with open(self.filename, "a") as file:
            file.write(formatted_entry)
        print("Done! Entry add successfully.")
    except PermissionError:
        print("Error: Permission denied.")

# This line 'teaches' the Librarian this skill
JournalManager.add_entry = add_entry

```

```python
def view_all_entries(self):
    try:
        with open(self.filename, "r") as file:
            content = file.read()
            if not content:
                print("\nYour journal is empty.")
            else:
                print("\n--- Your Journal ---\n" + content)
    except FileNotFoundError:
        print("\nJournal file not found. Create one by adding an entry!")

JournalManager.view_all_entries = view_all_entries
```

```python
def search_entry(self):
    keyword = input("\nEnter a keyword to search for: ").lower()
    try:
        with open(self.filename, "r") as file:
            entries = file.read().split('-'*30)
            found = False
            for entry in entries:
                if keyword in entry.lower() and entry.strip():
                    print(entry.strip() + f"\n{'-'*30}")
                    found = True
            if not found:
                print(f"No matches for '{keyword}'.")
    except FileNotFoundError:
        print("\nNo file to search.")

JournalManager.search_entry = search_entry
```

```python
def delete_all_entries(self):
    confirm = input("\nErase everything? (yes/no): ").lower()
    if confirm == 'yes':
        if os.path.exists(self.filename):
            os.remove(self.filename)
            print("Journal deleted.")
        else:
            print("Nothing to delete.")

JournalManager.delete_all_entries = delete_all_entries
```

```python
def start_app():
    bot = JournalManager()
    
    while True:
        print("\n--- MENU: 1:Add a new entry | 2:View all entry | 3:Search for an entry | 4:Delete all | 5:Exit ---")
        choice = input("Choice: ")
        
        if choice == '1': bot.add_entry()
        elif choice == '2': bot.view_all_entries()
        elif choice == '3': bot.search_entry()
        elif choice == '4': bot.delete_all_entries()
        elif choice == '5':
            print("Thank you ! Closing Journal. Goodby!")
            break
        else:
            print("Invalid Option.")

start_app()
```
